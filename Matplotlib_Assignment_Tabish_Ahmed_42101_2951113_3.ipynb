# Exerxise Simple Bar Chart II

```python
drinks = ['cappuccino', 'rum', 'cola', 'juice', 'ginger beer', 'vodka']
sales = [91, 76, 56, 66, 52, 27]
ax = plt.subplot()
ax.set_xticks(range(len(drinks)))
ax.set_xticklabels(["cappuccino", "latte", "chai", "americano", "mocha", "espresso"],rotation=30)
plt.bar(range(len(drinks)), sales)
plt.show()
```

```python
import numpy as np
drinks = ['cappuccino', 'rum', 'cola', 'juice', 'ginger beer', 'vodka']
sales1 = [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]
X_axis = np.arange(len(drinks))
  
plt.bar(X_axis - 0.2, sales1, 0.4, label = 'Sales1')
plt.bar(X_axis + 0.2, sales2, 0.4, label = 'Sales2')
  
plt.xticks(X_axis, drinks)
plt.xlabel("Drinks")
plt.ylabel("Number of Sales")
plt.title("Number of Sale for each drink")
plt.legend()
plt.show()
```

```python
drinks = ['cappuccino', 'rum', 'cola', 'juice', 'ginger beer', 'vodka']
sales1 = [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]
plt.bar(drinks, sales1, color='r',label = 'Location1')
plt.bar(drinks, sales2, bottom=sales1, color='b',label = 'Location2')
plt.legend()
plt.show()
```
